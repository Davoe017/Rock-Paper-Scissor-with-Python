```python
import random

while True:
    choices = ["rock", "paper", "scissors"]
    CPU = random.choice(choices)
    player = None
    print("Enter rock, paper, scissors")
    while player not in choices:
        player = input("rock, paper, or scissors?: ").lower()
        print("Enter rock, paper, scissors")
    if player == CPU:
        print("CPU: ", CPU)
        print("player: ", player)
        print("Game drawn!!!")
    elif player == "rock":
        if CPU == "scissors":
            print("CPU: ", CPU)
            print("player: ",  player)
            print("You win!!!")
        if CPU == "paper":
            print("CPU: ", CPU)
            print("player: ", player)
            print("You lose!!!")
    elif player == "scissors":
        if CPU == "paper":
            print("CPU: ", CPU)
            print("player: ", player)
            print("You win!!!")
        if CPU == "rock":
            print("CPU: ", CPU)
            print("player: ", player)
            print("You lose!!!")
    elif player == "paper":
        if CPU == "rock":
            print("CPU: ", CPU)
            print("player: ", player)
            print("You win!!!")
        if CPU == "scissors":
            print("CPU: ", CPU)
            print("player:", player)
            print("You lose!!!")
    keep_playing = input("Do you want to keep playing? (y for yes, press any key if no): ").lower()
    if keep_playing != "y":
        break
print("Game Over")
```
